### Agora que foram definidos os formatos de entrada e da pré-configuração, é preciso agora explicar como é o processo de montagem dos dataframes esperados pelo código.

Abaixo, você consegue visualizar como são escolhidas as colunas, a partir das medidas, relativas a um projeto.

```python
import pandas as pd
metrics = pre_config["measures"]
df_columns = metrics + ["qualifier"]
df = pd.DataFrame(columns=df_columns)
display(df.head())
```

### Uma vez definidas as colunas, é preciso popular os dados 

```python
for component in components:
    df.at[component["path"], "qualifier"] = component["qualifier"]
    for metric in component["measures"]:
        df.at[component["path"], metric["metric"]] = metric["value"]

print(f'Número de linhas do dataframe: {df.shape[0]}\n\n')
display(df.head())
```

### Uma vez compreendido o processo de popular o dataframe, agora vamos entender sobre o processo de filtrar o dataframe a partir do tipo de componente analisado pelo sonar (se é um diretório ou se é um arquivo, por exemplo)

Abaixo você pode visualizar um exemplo de dataframe filtado para componentes de diretório.

```python
dirs = df[df["qualifier"] == "DIR"]
print(f'Número de componentes diretório: {dirs.shape[0]}\n\n')
display(dirs)
```

### Agora vamos entender o funcionamento do retorno do idxmax, também utilizado no projeto. O idxmax é uma função no panda para retornar o índice do valor máximo de uma determinada coluna.

Abaixo, você primeiro pode verificar a coluna "tests", disponível no dataframe.

Logo abaixo dessa coluna, você pode verificar o resultado do idxmax a partir das seguintes operações:   
1. Separação da coluna "tests" 
2. Casting dos valores da pd.Series para float
3. Chamada da função idxmax para retornar os valores máximos
4. Aplicação do resultado das linhas anteriores para a linha "tests" do dataframe.

```python
print("Coluna tests:")
display(dirs["tests"])
print("\n")

print("Linha tests:")
display(dirs.loc["tests"])
print("\n")

print("Resultado das operações descritas acima:")
display(dirs.loc[dirs["tests"].astype(float).idxmax()])
print("\n")
```

### Para clarificar um pouco mais, segue um exemplo da utilização do idxmax

```python
example = pd.DataFrame({'consumption': [10.51, 103.11, 55.48],
                   'co2_emissions': [37.2, 19.66, 1712]},
                   index=['Pork', 'Wheat Products', 'Beef'])
print("DATAFRAME:")
display(example)
print("\n\n")

print("RESULTADO:")
display(example.idxmax())
print("\n\n")
```
